```python

# Initialize tokenizer and lemmatizer
tokenizer = RegexpTokenizer(r"\w+")
lemmatizer = WordNetLemmatizer()

# Get the English stopwords
stop_words = set(stopwords.words('english'))

# Function to lemmatize a given text and remove stopwords
def lemmatize_text_without_stopwords(text):
    lemmatized_words = [lemmatizer.lemmatize(word) for word in tokenizer.tokenize(text.lower()) if word not in stop_words]
    return lemmatized_words

# Apply lemmatization and remove stopwords to the 'tweet' column
df['lemmatized_tweet'] = df['tweet'].apply(lemmatize_text_without_stopwords)

# Function to convert tokens to text
def tokens_to_text(tokens):
    return ' '.join(tokens)

# Apply the function to create 'lemmatized_text_without_stopwords' column
df['lemmatized_text'] = df['lemmatized_tweet'].apply(tokens_to_text)

# Print the lemmatized text without stopwords
print(df['lemmatized_text'])
```

```python
text_data = smaller_df['lemmatized_text']

# Initialize the TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the text data using the vectorizer
tfidf_matrix = tfidf_vectorizer.fit_transform(text_data)

print(tfidf_matrix)
```

```python
# Convert the TF-IDF matrix to a Pandas DataFrame if needed
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
print(tfidf_df )
```

```python
#random_state fixes the train samples and test samples 
 # Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(tfidf_matrix, y, test_size=0.2, random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python


def evaluate_model(model, X_train, y_train, X_test, y_test):
    # Train the model
    model.fit(X_train, y_train)

    # Make predictions on training and testing data
    train_pred = model.predict(X_train)
    test_pred = model.predict(X_test)

    # Display model name
    print(type(model).__name__)

    # Evaluate on training data
    train_classification_report = classification_report(y_train, train_pred)
    print("Train Classification Report:")
    print(train_classification_report)
    ConfusionMatrixDisplay.from_estimator(model, X_train, y_train)
    plt.show()

    # Evaluate on testing data
    test_classification_report = classification_report(y_test, test_pred)
    print("Test Classification Report:")
    print(test_classification_report)
    ConfusionMatrixDisplay.from_estimator(model, X_test, y_test)
    plt.show()

models = [KNeighborsClassifier(), DecisionTreeClassifier(), RandomForestClassifier(), SVC()]

for model in models:
    evaluate_model(model, X_train, y_train, X_test, y_test)
```

```python

# Assuming you have the following lists
f1_score_train = [0.95, 1, 1, 0.97]
f1_score_test = [0.95, 0.94, 0.95, 0.94]
models = ['KNN', 'DecisionTree', 'RandomForest', 'SVC']

# Create an array of indices for the models
indices = np.arange(len(models))

# Set the width of the bars
bar_width = 0.35

# Create bar graphs for both training and test sets
plt.bar(indices, f1_score_train, bar_width, label='Training Set')
plt.bar(indices + bar_width, f1_score_test, bar_width, label='Test Set')

# Add labels, title, and legend
plt.xlabel('Models')
plt.ylabel('F1 Score')
plt.title('F1 Scores for Different Models')
plt.xticks(indices + bar_width / 2, models,fontsize=10)
plt.legend()

# Show the plot
plt.show()
```

```python
conf_matrix_KNeighborsClassifier = np.array([[465, 0], [25, 10]])
conf_matrix_DecisionTreeClassifier=np.array([[457,8],[20,15]])
conf_matrix_RandomForestClassifier=np.array([[465,0],[27,8]])
conf_matrix_SVC=np.array([[465,0],[31,4]])

TN1, FP1, FN1, TP1 = conf_matrix_KNeighborsClassifier.ravel()
TN2, FP2, FN2, TP2 = conf_matrix_DecisionTreeClassifier.ravel()
TN3, FP3, FN3, TP3 = conf_matrix_RandomForestClassifier.ravel()
TN4, FP4, FN4, TP4 = conf_matrix_SVC.ravel()

y_true_KNeighborsClassifier = [0] * TN1 + [0] * FP1 + [1] * FN1 + [1] * TP1
y_pred_KNeighborsClassifier= [0] * (TN1 + FP1) + [1] * (FN1 + TP1)

y_true_DecisionTreeClassifier = [0] * TN2 + [0] * FP2 + [1] * FN2 + [1] * TP2
y_pred_DecisionTreeClassifier= [0] * (TN2 + FP2) + [1] * (FN2 + TP2)

y_true_RandomForestClassifier = [0] * TN3 + [0] * FP3+ [1] * FN3 + [1] * TP3
y_pred_RandomForestClassifier= [0] * (TN3 + FP3) + [1] * (FN3 + TP3)

y_true_SVC = [0] * TN4 + [0] * FP4+ [1] * FN4 + [1] * TP4
y_pred_SVC= [0] * (TN4 + FP4) + [1] * (FN4 + TP4)

auc_KNeighborsClassifier= np.round(roc_auc_score(y_true_KNeighborsClassifier, y_pred_KNeighborsClassifier), 3)
auc_DecisionTreeClassifier= np.round(roc_auc_score(y_true_DecisionTreeClassifier, y_pred_DecisionTreeClassifier), 3)
auc_RandomForestClassifier= np.round(roc_auc_score(y_true_RandomForestClassifier, y_pred_RandomForestClassifier), 3)
auc_SVC= np.round(roc_auc_score(y_true_SVC, y_pred_SVC), 3)

print("AUC for KNeighborsClassifier is {}".format(auc_KNeighborsClassifier))
print("AUC for DecisionTreeClassifieris {}".format(auc_DecisionTreeClassifier))
print("AUC for RandomForestClassifier is {}".format(auc_RandomForestClassifier))
print("AUC for SVC is {}".format(auc_SVC))
```

```python
conf_matrix_KNeighborsClassifier = np.array([[465, 0], [27, 8]])
conf_matrix_DecisionTreeClassifier = np.array([[452, 13], [16, 19]])
conf_matrix_RandomForestClassifier = np.array([[465, 0], [29, 65]])
conf_matrix_SVC = np.array([[465, 0], [31, 4]])

# Extract true positive rate (TPR) and false positive rate (FPR) for each classifier
fpr_KNN, tpr_KNN, _ = roc_curve(y_true_KNeighborsClassifier, y_pred_KNeighborsClassifier)
fpr_DT, tpr_DT, _ = roc_curve(y_true_DecisionTreeClassifier, y_pred_DecisionTreeClassifier)
fpr_RF, tpr_RF, _ = roc_curve(y_true_RandomForestClassifier, y_pred_RandomForestClassifier)
fpr_SVC, tpr_SVC, _ = roc_curve(y_true_SVC, y_pred_SVC)

# Calculate AUC for each classifier
roc_auc_KNN = auc(fpr_KNN, tpr_KNN)
roc_auc_DT = auc(fpr_DT, tpr_DT)
roc_auc_RF = auc(fpr_RF, tpr_RF)
roc_auc_SVC = auc(fpr_SVC, tpr_SVC)

# Plot ROC curves
plt.figure(figsize=(8, 6))

plt.plot(fpr_KNN, tpr_KNN, label=f'KNeighborsClassifier (AUC = {roc_auc_KNN:.2f})')
plt.plot(fpr_DT, tpr_DT, label=f'DecisionTreeClassifier (AUC = {roc_auc_DT:.2f})')
plt.plot(fpr_RF, tpr_RF, label=f'RandomForestClassifier (AUC = {roc_auc_RF:.2f})')
plt.plot(fpr_SVC, tpr_SVC, label=f'SVC (AUC = {roc_auc_SVC:.2f})')

plt.plot([0, 1], [0, 1], 'k--', label='Random')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('ROC Curve for Different Classifiers')
plt.legend(loc='lower right')
plt.show()
```
